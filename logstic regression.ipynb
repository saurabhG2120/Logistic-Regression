```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
df["species"].unique()
```

```python
df.isnull().sum()
```

```python
#removing setosa as we are performing binary classification
df=df[df["species"]!="setosa"]
```

```python
df.head()
```

```python
#mapping "versicolor":0,"virginica":1
df["species"]=df["species"].map({"versicolor":0,"virginica":1})
```

```python
df.head()
```

```python
###splittig dataset into indipendent and dependent features
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameters={"penalty":["l1","l2","elasticnet"],"C":[1,2,3,4,5,6,10,20,30,40,50,60],
            "max_iter":[100,200,300,400]}
```

```python
classifire_reg=GridSearchCV(classifier,param_grid=parameters,
                            scoring="accuracy",cv=5)
```

```python
classifire_reg.fit(X_train,y_train)
```

```python
print(classifire_reg.best_params_)
```

```python
print(classifire_reg.best_score_)
```

```python
classifire_reg.predict(X_test)
```

```python
##predicition
y_pred=classifire_reg.predict(X_test)
```

```python
#accuracy score
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
